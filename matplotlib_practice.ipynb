```python
import matplotlib.pyplot as plt
```

```python
#sets the x-axis as strings for months
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#sets y-axis as float for the toal fare in US dollars accumulated each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
plt.plot(x_axis,y_axis)
plt.xlabel('Month')
plt.ylabel('Savings ($)')
plt.title('Saving Account Results')
plt.show
```

```python
#create the plot with ax.plt()

fig, ax = plt.subplots()

ax.plot(x_axis,y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
ax = plt.axes()
ax.plot(x_axis,y_axis)
```

```python
plt.plot(x_axis,y_axis, marker="*",color="blue",linewidth=2, label='Boston')

plt.xlabel("Date")
plt.ylabel("Fare($)")
plt.ylim(0,45)
plt.xlim(0,5)
#plt.figure(figsize=("20","10"))
plt.title("PyBer Fare by Month")
plt.legend()
```

```python

ax = plt.axes()
ax.plot(x_axis,y_axis,c="g",marker="D",linewidth=2,label="Boston")
ax.legend()
ax.set_xlabel("Month")
ax.set_ylabel("Fare($)")
ax.set_title("PyBer Fare by Month")
ax.set_xlim(0,13)
ax.set_ylim(0,45)
ax.grid()
```

```python
# Set the x-axis to a list of strings for each month.
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

# Set the y-axis to a list of floats as the total fare in US dollars accumulated for each month.
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#create the plot with ax.plt()

fig, ax = plt.subplots()

ax.barh(x_axis,y_axis,color="cyan",label="Boston")
ax.invert_yaxis()
ax.legend()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer Fare by Month")
```

```python
plt.plot(x_axis,y_axis,'o')
```

```python
#y_axis_larger = []

#for data in y_axis:
  #  y_axis_larger.append(data*3)
    

plt.scatter(y_axis,x_axis, s = [i*3 for i in y_axis], label = "Chicago",color = "red")
plt.title("PyBer Fare by Month")
plt.ylabel("Month")
plt.xlabel("Date")
plt.legend()
plt.gca().invert_yaxis()
```

```python
fig, ax = plt.subplots(figsize=(8,8))

ax.scatter(y_axis,x_axis, s=[i*5 for i in y_axis], c ='skyblue',edgecolors="black",linewidth=2,alpha=0.2,label="Boston")
ax.legend()
ax.invert_yaxis()
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
ax.set_title("PyBer by Month")
ax.set_xlim(0,50)
```

```python
plt.pie(y_axis,labels=x_axis)
plt.show
```

```python
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]

plt.pie(y_axis, explode=explode_values, colors = colors, labels=x_axis, autopct='%.1f%%')
plt.subplots(figsize=(8, 8))
plt.show()
```
